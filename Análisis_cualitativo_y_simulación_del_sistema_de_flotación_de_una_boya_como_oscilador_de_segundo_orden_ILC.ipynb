## **Análisis cualitativo y simulación del sistema de flotación de una boya como oscilador de segundo orden**



Este trabajo analiza el movimiento vertical de una boya flotante, modelado mediante una ecuación diferencial ordinaria de segundo orden. El sistema se linealiza para pequeñas oscilaciones, lo que permite calcular su posición de equilibrio, frecuencia natural y estudiar su estabilidad. A través de simulaciones numéricas, se obtienen gráficas de la respuesta temporal y del espacio de estados, lo que facilita la interpretación del comportamiento dinámico.

Este tipo de análisis es relevante para el diseño de sistemas de generación de energía mareomotriz y undimotriz, donde las boyas se utilizan para convertir la energía de las olas en energía eléctrica. Comprender la dinámica de la boya permite optimizar su tamaño, masa y amortiguamiento para maximizar la captación de energía y mejorar la eficiencia del convertidor electromecánico acoplado. La analogía con un oscilador masa-resorte-amortiguador o un circuito RLC facilita la aplicación de técnicas de control y resonancia en ingeniería eléctrica para obtener el mayor aprovechamiento de la energía del oleaje.

# **Introducción**

La creciente necesidad de fuentes de energía renovable ha impulsado el desarrollo de tecnologías que aprovechan la energía del océano. Entre ellas, la generación mareomotriz y undimotriz se basa en convertir el movimiento de las mareas y las olas en energía eléctrica. Un componente esencial de muchos convertidores de energía de olas es la boya flotante, la cual se desplaza verticalmente siguiendo el movimiento de la superficie del agua.

Para diseñar sistemas eficientes es crucial comprender la dinámica de estas boyas. Su comportamiento puede modelarse mediante ecuaciones diferenciales ordinarias de segundo orden, análogas a las de un oscilador masa-resorte-amortiguador o a un circuito RLC en ingeniería eléctrica. Este modelo permite predecir la posición de equilibrio, la frecuencia natural de oscilación y evaluar la estabilidad del sistema. Estos resultados sirven de base para optimizar el acoplamiento mecánico y eléctrico del generador, buscando maximizar la transferencia de energía en condiciones reales de oleaje.

# **Descripción del fenómeno**

Consideremos una **boya cilíndrica** de altura \(h\) flotando en el mar.  
Denotemos por \(x(t)\) la altura que sobresale de la superficie del agua en función del tiempo.  
La boya está sometida principalmente a dos fuerzas:

- **Peso:**
$$P = mg, $$
fuerza gravitatoria que tiende a hundirla.
- **Empuje hidrostático:**
$$
E = \rho g A \bigl(h - x(t)\bigr),
$$proporcional al volumen de fluido desplazado.

Aplicando la **segunda ley de Newton** en dirección vertical y considerando desplazamientos pequeños alrededor de la posición de equilibrio, se obtiene la ecuación diferencial lineal:

$$
x''(t) + \frac{3g}{2h}x(t) - \frac{g}{2} = 0
$$

- El segundo término representa la **rigidez** del sistema (fuerza de restitución).
- El término constante fija la **posición de equilibrio**, dada por:

$$
x_e = \frac{h}{3}
$$

Alrededor de este equilibrio, la boya realiza oscilaciones armónicas con **frecuencia natural** y **periodo**:

$$
\omega_n = \sqrt{\frac{3g}{2h}}, \qquad
T = \frac{2\pi}{\omega_n}
$$

Este fenómeno es un ejemplo clásico de **oscilaciones libres** y puede extenderse añadiendo:
- **Amortiguamiento**, para representar pérdidas por rozamiento del agua.
- **Forzamiento periódico**, para modelar el efecto de olas u otras perturbaciones.

Estos elementos hacen que el modelo sea altamente representativo de la dinámica en dispositivos de **conversión de energía de olas**, donde la boya actúa como un oscilador que transfiere su energía a un generador electromecánico.


# **Modelo analítico y análisis cualitativo**

Partimos de la formulación simplificada del problema de la boya (ver enunciado). La ecuación resultante, para oscilaciones pequeñas, es:

$$
x''(t) + \frac{3g}{2h}\,x(t) - \frac{g}{2} = 0,
$$

donde \(x(t)\) es la altura que sobresale de la superficie, \(h\) la altura total de la boya y \(g\) la aceleración de la gravedad.

**Equilibrio.** Para el estado estacionario \(x''=0\) se tiene:
$$
\frac{3g}{2h}x_e - \frac{g}{2} = 0 \quad\Rightarrow\quad x_e=\frac{h}{3}.
$$

Definiendo la variable de desviación \(y(t)=x(t)-x_e\), la ecuación para \(y\) es homogénea:

$$
y''(t) + \omega_n^2\,y(t) = 0,\qquad \omega_n=\sqrt{\frac{3g}{2h}}.
$$

Esto es un **oscilador armónico simple** con frecuencia natural $\omega_n$ y periodo $(T=2\pi/\omega_n)$. La solución general es:

$$
y(t) = C_1\cos(\omega_n t) + C_2\sin(\omega_n t),
$$

y por tanto
$$
x(t)=x_e + C_1\cos(\omega_n t) + C_2\sin(\omega_n t).
$$

**Estabilidad.**  
Las soluciones son marginalmente estables (oscilaciones periódicas) en ausencia de disipación.  
Si añadimos un término de amortiguamiento lineal \(c\,x'(t)\), la ecuación pasa a:

$$
x''(t) + c\,x'(t) + \frac{3g}{2h}x(t) - \frac{g}{2} = 0,
$$

y el equilibrio se vuelve **asintóticamente estable** siempre que \(c>0\): las oscilaciones decaen con el tiempo.  
El **coeficiente crítico de amortiguamiento** que separa el régimen oscilatorio del no oscilatorio es:

$$
c_{\mathrm{crit}} = 2\sqrt{k}, \qquad k = \frac{3g}{2h}.
$$

- Para $c_{\mathrm{crit}}$: **subamortiguado** → oscilaciones decrecientes.  
- Para  $c_{\mathrm{crit}}$: **críticamente amortiguado** → decae sin oscilar (más rápido posible).  
- Para  $c_{\mathrm{crit}}$: **sobreamortiguado** → decae de forma monótona, sin oscilaciones.

**Interpretación para energía mareomotriz / undimotriz.**  
La frecuencia natural \(\omega_n\) informa sobre la selección de la geometría/masa de la boya para acoplarla a la banda energética del oleaje (buscar coincidencia con la frecuencia dominante de las olas para aprovechar resonancia controlada). El término de amortiguamiento modela pérdidas y la extracción de energía (por ejemplo, por el generador acoplado).


```python
# === Análisis simbólico y numérico del modelo (sympy) ===
# Si te falta sympy en Colab: descomenta e instala
# !pip install sympy

import sympy as sp
sp.init_printing()

# símbolos
g, h = sp.symbols('g h', positive=True)
omega2 = 3*g/(2*h)

# equilibrio
x_e = sp.simplify((g/2) / omega2)   # (g/2) / (3g/(2h)) = h/3
omega = sp.simplify(sp.sqrt(omega2))
T = sp.simplify(2*sp.pi/omega)

# mostrar
x_e, omega, T
```

# **Modelo computacional y simulaciones**

Aquí implementamos simulaciones numéricas para:

1. Comparar la **solución analítica** (sin amortiguamiento) con la solución numérica (`solve_ivp`).
2. Graficar el **espacio de estados** \(x\) vs \(x'\).
3. Estudiar el **efecto del amortiguamiento** (subamortiguado / crítico / sobreamortiguado).
4. Simular la respuesta forzada si se desea (p. ej. olas: forzamiento armónico \(F(t)=A\cos(\omega t)\)).
5. Hacer un estudio de parámetros (p. ej. variar la altura `h` y observar \(\omega_n\)).

Las celdas de código siguientes ejecutan cada parte. Ajusta parámetros (`h_val`, amortiguamiento `c`, amplitud/ frecuencia de forzamiento) según necesites.


```python
# --- 1) Comparación analítica vs numérica (sin amortiguamiento) ---

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parámetros
g_val = 9.81
h_val = 2.0
omega_n = np.sqrt(3*g_val/(2*h_val))
x_eq = h_val/3.0

# Solución analítica
def x_analytic(t, h=h_val):
    w = np.sqrt(3*g_val/(2*h))
    return -h/3.0 * np.cos(w*t) + h/3.0

# Ecuación diferencial sin amortiguamiento
def f_free(t, X, g=g_val, h=h_val):
    return [X[1], - (3*g/(2*h))*X[0] + g/2.0]

# Resolución numérica
t_span = (0.0, 6.0)
t_eval = np.linspace(*t_span, 2000)
X0 = [0.0, 0.0]
sol = solve_ivp(lambda t,y: f_free(t,y), t_span, X0, t_eval=t_eval, rtol=1e-8, atol=1e-10)

# Gráfica
plt.figure(figsize=(10,4))
plt.plot(sol.t, sol.y[0], label='Numérica')
plt.plot(t_eval, x_analytic(t_eval), '--', label='Analítica')
plt.axhline(x_eq, color='k', linestyle=':', alpha=0.5, label='Equilibrio')
plt.xlabel('t (s)')
plt.ylabel('x (m)')
plt.title('Comparación analítica vs numérica (sin amortiguamiento)')
plt.legend()
plt.grid(True)
plt.show()
```

La curva numérica (línea continua) y la analítica (línea discontinua) coinciden, mostrando que el integrador numérico reproduce correctamente la solución exacta. La gráfica es sinusoidal, lo que confirma que el sistema es un oscilador armónico simple que oscila alrededor de la línea de equilibrio x.

```python
# --- Espacio de estados ---
plt.figure(figsize=(6,6))
plt.plot(sol.y[0], sol.y[1])
plt.xlabel('x (m)')
plt.ylabel("x' (m/s)")
plt.title('Espacio de estados (sin amortiguamiento)')
plt.grid(True)
plt.show()
```

El resultado es una elipse cerrada, característica de un oscilador no amortiguado. La trayectoria indica que la energía total del sistema se conserva (no hay fricción), por eso no hay espiral sino una curva cerrada.

```python
#Efecto del amortiguamiento
k = 3*g_val/(2*h_val)
c_crit = 2*np.sqrt(k)
c_values = [0.1*c_crit, c_crit, 5*c_crit]
labels = ['Subamortiguado', 'Crítico', 'Sobreamortiguado']

def simulate_with_damping(c_val, t_span=(0,6), X0=[0.0,0.0], t_points=2000):
    def f_damp(t, X):
        return [X[1], -c_val*X[1] - k*X[0] + g_val/2.0]
    t_eval = np.linspace(t_span[0], t_span[1], t_points)
    sol = solve_ivp(f_damp, t_span, X0, t_eval=t_eval)
    return sol.t, sol.y

plt.figure(figsize=(10,6))
for c_val, lab in zip(c_values, labels):
    t_arr, y_arr = simulate_with_damping(c_val)
    plt.plot(t_arr, y_arr[0], label=f'{lab}, c={c_val:.3f}')
plt.axhline(x_eq, color='k', linestyle=':', alpha=0.5)
plt.xlabel('t (s)')
plt.ylabel('x (m)')
plt.title('Respuesta para distintos amortiguamientos')
plt.legend()
plt.grid(True)
plt.show()
```


Para subamortiguado, hay oscilaciones decrecientes (espiras que se achican con el tiempo).

Para críticamente amortiguado, el sistema vuelve al equilibrio lo más rápido posible sin oscilar.

Para sobreamortiguado, el sistema regresa de forma lenta y monótona al equilibrio.

Esto demuestra que el parámetro c controla la disipación de energía y, por lo tanto, el tiempo de asentamiento del sistema.

```python
#Respuesta forzada y barrido de parámetros
A = 0.1
omega_forcing = omega_n * 0.9
c_example = 0.2*c_crit

def f_forced(t, X):
    return [X[1], -c_example*X[1] - k*X[0] + g_val/2.0 + A*np.cos(omega_forcing*t)]

sol_forced = solve_ivp(f_forced, t_span, X0, t_eval=t_eval)
plt.figure(figsize=(10,4))
plt.plot(sol_forced.t, sol_forced.y[0])
plt.xlabel('t (s)')
plt.ylabel('x (m)')
plt.title(f'Respuesta forzada: A={A}, ω_f={omega_forcing:.3f} rad/s')
plt.grid(True)
plt.show()

# --- 4) Barrido de parámetros ---
h_values = np.linspace(0.5, 5.0, 20)
omega_vals = np.sqrt(3*g_val/(2*h_values))
plt.figure(figsize=(8,4))
plt.plot(h_values, omega_vals, 'o-')
plt.xlabel('h (m)')
plt.ylabel('ω_n (rad/s)')
plt.title('Frecuencia natural vs altura de la boya h')
plt.grid(True)
plt.show()
```

La respuesta forzada muestra una oscilación estacionaria cuya amplitud depende de la frecuencia de excitación: si está cerca de $w_n$ la amplitud es mayor (fenómeno de resonancia).
El barrido de parámetros muestra que la frecuencia natural disminuye a medida que la boya es más alta (mayor masa efectiva → oscilaciones más lentas).

# **Análisis de resultados y conclusiones**

Los resultados obtenidos muestran que:

- La **solución analítica** sin amortiguamiento coincide con la solución numérica, validando el modelo.
- La **posición de equilibrio** se encuentra en:

$$
x_e = \frac{h}{3}
$$

lo que es coherente con la condición de empuje igual al peso.

- La **frecuencia natural** del sistema es:

$$
\omega_n = \sqrt{\frac{3g}{2h}}
$$

Por lo tanto, al aumentar la altura \(h\) de la boya, la frecuencia disminuye, produciendo oscilaciones más lentas.

- La introducción de **amortiguamiento** provoca que las oscilaciones decaigan en el tiempo. En el régimen subamortiguado, el sistema sigue oscilando pero con amplitud decreciente; en el régimen críticamente amortiguado el sistema regresa al equilibrio en el menor tiempo sin oscilar; y en el sobreamortiguado no se presentan oscilaciones, sino un retorno monótono al equilibrio.
- La adición de **forzamiento armónico** permite estudiar condiciones de resonancia. Si la frecuencia de forzamiento se aproxima a $\omega_n $, la amplitud de oscilación aumenta. Esto es fundamental en el diseño de sistemas de energía undimotriz para aprovechar la resonancia y maximizar la captura de energía.
