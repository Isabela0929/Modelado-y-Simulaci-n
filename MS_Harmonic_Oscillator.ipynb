```python
from scipy.integrate import solve_ivp

def Harmonic_Oscillator_DS (t,z,k,m):
  dzdt = [z[1], -k/m*z[0]]
  return dzdt

t_span = (0,10)
z0 = [1,0]
k = 1
m = 1

z_sol = solve_ivp(Harmonic_Oscillator_DS, t_span, z0, args=(k,m), dense_output=True);
```

```python
print(z_sol)
```

```python
import matplotlib.pyplot as plt
import numpy as np

tplot= np.linspace (0,10,100)
zplot = z_sol.sol(tplot)

plt.plot(tplot,zplot[0])
plt.plot(tplot,zplot[1])
plt.show()
```
